```python
import sqlite3 as s

conn = s.connect('./db/MembersDB.db')
c = conn.cursor()
```

```python
c.execute('drop table if exists Address')
c.execute('create table address(depart  char(100), name char(30), phone char(50), age int)')
c.execute('insert into address values("동료", "김연아", "010-9872-0202", 22)')
c.execute('insert into address values("친구", "박태환", "010-8585-7896", 25)')
c.execute('insert into address values("친구", "이승만", "010-7852-1258", 25)')
c.execute('insert into address values("동료", "이순신", "010-6547-1234", 55)')
c.execute('insert into address values("동호회", "강감찬", "010-8523-9872", 35)')
c.execute('insert into address values("기타", "세탁소", "02-111-2222", 12)')
c.execute('insert into address(name, phone) values("세종대왕", "010-8523-9872")')
c.execute('insert into address(name, phone, depart) values("치킨집", "02-3445-2222", "기타")')

conn.commit
```

```python
# 메뉴판 테이블 생성 및 데이터 입력
c.execute('drop table if exists MenuPan')
c.execute('create table MenuPan(menuName char(50), price int)')
c.execute('insert into MenuPan values("김밥", 3000)')
c.execute('insert into MenuPan values("순두부", 6000)')
c.execute('insert into MenuPan values("육개장", 7500)')
c.execute('insert into MenuPan values("칼국수", 5500)')
c.execute('insert into MenuPan values("짜장밥", 5500)')
c.execute('insert into MenuPan values("김치찌개",8000)')
c.execute('insert into MenuPan(price, menuName) values(800,"aaa")')
c.execute('insert into MenuPan(price, menuName) values(666,"bbb")')

conn.commit()
```

```python
def menuPanSelect():
    c.execute('select * from menuPan')
    rows = c.fetchall()

    for row in rows:
        print("{} {}".format(row[0], row[1]))
```
